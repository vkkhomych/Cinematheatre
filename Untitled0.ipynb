```python
import pandas as pd
data = pd.read_csv('/content/data.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data.describe()
```

```python
train1 = data.drop(['Credit'],axis=1)
```

```python
train1.head()
```

```python
job2int = {job: num  for num, job in enumerate(list(set(train1['Job Family'])))}
train1['Job Family'] = train1['Job Family'].map(lambda x : job2int[x])

job2int = {job: num  for num, job in enumerate(list(set(train1['Gender Key'])))}
train1['Gender Key'] = train1['Gender Key'].map(lambda x : job2int[x])

job2int = {job: num  for num, job in enumerate(list(set(train1['City'])))}
train1['City'] = train1['City'].map(lambda x : job2int[x])

job2int = {job: num  for num, job in enumerate(list(set(train1['Marital Status'])))}
train1['Marital Status'] = train1['Marital Status'].map(lambda x : job2int[x])

job2int = {job: num  for num, job in enumerate(list(set(train1['MBO Eligible'])))}
train1['MBO Eligible'] = train1['MBO Eligible'].map(lambda x : job2int[x])

job2int = {job: num  for num, job in enumerate(list(set(train1['Full time?'])))}
train1['Full time?'] = train1['Full time?'].map(lambda x : job2int[x])
```

```python
train1.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
labels = data['Credit']
```

```python
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.25,random_state =2)
```

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 30, max_depth = 10, min_samples_split = 5,
          learning_rate = 0.1, loss = 'ls')
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test,y_test)
```

```python
y_pred = clf.predict(x_test)
```
